```python
# Merge the ALI with the MSAG sync file 
# The purpose is to see how many ALI entries are locking the PSAP out of editing  specific MSAG entries- 
# Outer merge with ALI on left. 
```

```python
# REQUIRED 
import pandas as pd
```

```python
# import the ALI file and the MSAG sync file

ali_path = (r"path/to/ali.csv")

msag_sync_path = (r"path/to/msag-ssap/sync/file.csv")


ali = pd.read_csv(ali_path)

sync_file = pd.read_csv(msag_sync_path)
```

```python
# print(ali)
```

```python
# print(sync_file)
```

```python
# Merging with the ALI as the left dataframe
# Using Street name, Community name, and street prefix directional as the keys. Each MSAG entry may have more than one ALI entry due to MSAG range vs ALI individual records

ali_match = pd.merge(ali, sync_file, left_on=['STREET_PREFIX_DIRECTIONAL','STREET_NAME', 'COMMUNITY_NAME' ], right_on=['STREET_PREFIX_DIRECTIONAL','STREET_NAME','COMMUNITY_NAME'], how='outer', indicator='ALI_merge')


# update the indicator messages
message = {"left_only":"ALI only", "both":"both", "right_only":"MSAG sync only"}
ali_match['ALI_merge'] = ali_match['ALI_merge'].map(message)
```

```python
ali_match.to_csv(r"path/to/export.csv")
```
